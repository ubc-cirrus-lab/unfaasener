```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
timing_df = pd.read_csv('./Timing_df.csv', index_col=0)
timing_df = timing_df.groupby(['n_funcs', 'n_hosts', 'sol']).median().reset_index()
```

```python
plt_df = timing_df[['n_funcs', 'n_hosts', 'time']]
plt_df = plt_df.pivot(index='n_funcs', columns='n_hosts', values='time')
sns.heatmap(plt_df, annot=False, cmap='YlGnBu', fmt='g')
```

```python
%matplotlib widget
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(
    timing_df['n_funcs'],
    timing_df['n_hosts'],
    timing_df['time']
)
```
