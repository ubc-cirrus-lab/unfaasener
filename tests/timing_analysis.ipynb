```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
timing_df = pd.read_csv('./Timing_df_Gekko.csv', index_col=0)
timing_df.loc[timing_df['sol_gekko'] == 'NotFound', 'time_gekko'] = 120
timing_df = timing_df.groupby(['n_funcs', 'n_hosts', 'sol_gekko', 'sol_julia']).median().reset_index()
```

```python
timing_df
```

```python
plt_df_julia = timing_df[['n_funcs', 'n_hosts', 'time_julia']]
plt_df_julia = plt_df_julia.pivot(index='n_funcs', columns='n_hosts', values='time_julia')
sns.heatmap(plt_df_julia, annot=True, cmap='YlGnBu', fmt='g', vmin=0, vmax=120)
```

```python
plt_df_gekko = timing_df[['n_funcs', 'n_hosts', 'time_gekko']]
plt_df_gekko = plt_df_gekko.pivot(index='n_funcs', columns='n_hosts', values='time_gekko')
sns.heatmap(plt_df_gekko, annot=True, cmap='YlGnBu', fmt='g', vmin=0, vmax=120)
```

```python
timing_df['time_rel'] = np.divide(timing_df['time_gekko'], timing_df['time_julia'])
plt_df_rel = timing_df[['n_funcs', 'n_hosts', 'time_rel']]
plt_df_rel = plt_df_rel.pivot(index='n_funcs', columns='n_hosts', values='time_rel')
sns.heatmap(plt_df_rel, annot=True, cmap='YlGnBu', fmt='g')
```

```python
%matplotlib widget
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(
    timing_df['n_funcs'],
    timing_df['n_hosts'],
    timing_df['time']
)
```
